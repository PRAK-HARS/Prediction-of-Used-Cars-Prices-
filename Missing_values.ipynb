```python
import copy
df_train = copy.deepcopy(dataset)
```

```python
df_train.head()
```

```python
df_train['Horse_power']= df_train['engine'].str.extract(r'(\d+)\.?\d*HP').astype(float)
df_train['Engine_Displacement']= df_train['engine'].str.extract(r'(\d+\.?\d*)L').astype(float)
df_train['Cylinder_Count']= df_train['engine'].str.extract(r'(\d+) Cylinder ').astype(float)
```

```python
df_train.columns
```

```python
for col in df_train.columns:    
    null_count = df_train[col].isna().sum()

# Output the result
    print(f"Number of null values in {col}: {null_count}")
```

```python
for index, row in df_train[df_train['Horse_power'].isna()].iterrows():
        # Get the car model with missing horse power
        car_model = row['model']
        
        # Find the median horse power for the same car model (excluding NaN values)
        median_hp = df_train[df_train['model'] == car_model]['Horse_power'].median()

        # Replace NaN with the median horse power of that car model
        df_train.at[index, 'Horse_power'] = median_hp

# Fill the missing horse power values
```

```python
for col in df_train.columns:    
    null_count = df_train[col].isna().sum()

# Output the result
    print(f"Number of null values in {col}: {null_count}")
```

```python
for index, row in df_train[df_train['Engine_Displacement'].isna()].iterrows():
        # Get the car model with missing horse power
        car_model = row['model']
        
        # Find the median horse power for the same car model (excluding NaN values)
        median_hp = df_train[df_train['model'] == car_model]['Engine_Displacement'].median()

        # Replace NaN with the median horse power of that car model
        df_train.at[index, 'Engine_Displacement'] = median_hp

# Fill the missing horse power values
```

```python
for index, row in df_train[df_train['Cylinder_Count'].isna()].iterrows():
        # Get the car model with missing horse power
        car_model = row['model']
        
        # Find the median horse power for the same car model (excluding NaN values)
        median_hp = df_train[df_train['model'] == car_model]['Cylinder_Count'].median()

        # Replace NaN with the median horse power of that car model
        df_train.at[index, 'Cylinder_Count'] = median_hp

# Fill the missing horse power values
```

```python
cylinder_config={'V','Flat','Straight'}
Turbocharger = {'Turbo','Twin Turbo'}
```

```python
def zoo(word,cylinder_config):
  for i in cylinder_config:
    if i in word:
      return i
  return 'Nan'
```

```python

df_train['cylinder_config']=df_train['engine'].apply(zoo,args=(cylinder_config,))
df_train['Turbocharger']=df_train['engine'].apply(zoo,args=(Turbocharger,))
```
